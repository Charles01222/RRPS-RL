```python
class BotAgent(rl_agent.AbstractAgent):
    """Agent class that wraps a bot."""
    def __init__(self, num_actions, bot, name="bot_agent"):
        assert num_actions > 0
        self._bot = bot
        self._num_actions = num_actions

    def restart(self):
        self._bot.restart()

    def step(self, time_step, is_evaluation=False):
        if time_step.last():
            return
        _, state = pyspiel.deserialize_game_and_state(
            time_step.observations["serialized_state"])
        action = self._bot.step(state)
        probs = np.zeros(self._num_actions)
        probs[action] = 1.0
        return rl_agent.StepOutput(action=action, probs=probs)

def eval_agents(env, agents, num_players, num_episodes):
    """Evaluate the agent."""
    sum_episode_rewards = np.zeros(num_players)
    for ep in range(num_episodes):
        for agent in agents:
            if hasattr(agent, "restart"):
                agent.restart()
        time_step = env.reset()
        episode_rewards = np.zeros(num_players)
        while not time_step.last():
            agents_output = [agent.step(time_step, is_evaluation=True) for agent in agents]
            action_list = [agent_output.action for agent_output in agents_output]
            time_step = env.step(action_list)
            episode_rewards += time_step.rewards
        sum_episode_rewards += episode_rewards
    return sum_episode_rewards / num_episodes

def create_roshambo_bot_agent(player_id, num_actions, bot_names, pop_id):
    name = bot_names[pop_id]
    bot = pyspiel.make_roshambo_bot(player_id, name)
    return BotAgent(num_actions, bot, name=name)

# 配置环境
num_episodes = 1
env_recall = 1
env = rl_environment.Environment(
    "repeated_game(stage_game=matrix_rps(),num_repetitions=" +
    f"{pyspiel.ROSHAMBO_NUM_THROWS}," +
    f"recall={env_recall})",
    include_full_state=True)

num_players = 2
num_actions = env.action_spec()["num_actions"]

# 加载Roshambo机器人
roshambo_bot_names = pyspiel.roshambo_bot_names()
roshambo_bot_names.sort()

# 准备记录所有对战结果的DataFrame
columns = ["Player 0", "Player 1", "Player 0 Avg Returns", "Player 1 Avg Returns"]
results_list = []

print(f"Total number of bots is {len(roshambo_bot_names)}.")

# 让每个机器人两两对战
for i in range(len(roshambo_bot_names)):
    for j in range(len(roshambo_bot_names)):
        # 创建两个机器人
        agents = [
            create_roshambo_bot_agent(0, num_actions, roshambo_bot_names, i),
            create_roshambo_bot_agent(1, num_actions, roshambo_bot_names, j)
        ]

        # 运行对抗赛
        avg_eval_returns = eval_agents(env, agents, num_players, num_episodes)

        # 记录结果
        results_list.append({
            "Player 0": roshambo_bot_names[i],
            "Player 1": roshambo_bot_names[j],
            "Player 0 Avg Returns": avg_eval_returns[0],
            "Player 1 Avg Returns": avg_eval_returns[1]
        })
    print(f"Bot no.{i}: {roshambo_bot_names[i]} has finished its competitions against all the bots.")

# 将列表转换为DataFrame
results_df = pd.DataFrame(results_list, columns=columns)

# 展示最终的对战结果
print(results_df)
```

```python
pivoted = results_df.pivot(index='Player 0', columns='Player 1', values='Player 0 Avg Returns')
row_sum =  pivoted.sum(axis=1)
pivoted_sorted = pivoted.loc[row_sum.sort_values(ascending=False).index,row_sum.sort_values(ascending=False).index]
pivoted_sorted
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 16))
ax1 = sns.heatmap(pivoted_sorted, annot=False, cmap="coolwarm", linewidths=.1)
ax1.set(xlabel=None)
ax1.set(ylabel=None)
plt.title('Heatmap of RoShamBo bots playoff')
plt.show()
```
